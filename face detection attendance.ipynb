```python
import face_recognition as fr
from PIL import Image,ImageDraw
class_gp=fr.load_image_file("group.jpg")
student_1=fr.load_image_file("student1.jpg")
student_2=fr.load_image_file("student2.jpg")
```

```python
locations=fr.face_locations(class_gp)
encodeings=fr.face_encodings(class_gp)

print("Total Students Present:",len(locations))
if len(locations)==0:
  print("no students were present")
```

```python
students={"student_1": fr.face_encodings(student_1),
"student_2":fr.face_encodings(student_2)}
```

```python
from PIL import Image as PIL_Image
attendance={}
present=0

image= PIL_Image.fromarray(class_gp)
draw= ImageDraw.Draw(image)
```

```python
for name,known_encoding in students.items():
  found= False

  for i, encoding in enumerate(encodeings):
    match=fr.compare_faces([known_encoding],encoding)
    if match[0].any():
      found= True
      present+=1
      top, right, bottom, left = locations[i]
      draw.rectangle(((left, top), (right, bottom)), outline="green",width=4)

      break
  attendance[name]=found

print("--Atendance Report--")

for name, status in attendance.items():
  print(name,"-","Present"if status else "Person not from the class")
print(f"Total present: {present}/{len(locations)}")

image.save("attendance.jpg")
image.show()
```
