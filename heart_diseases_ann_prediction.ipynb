```python
import pandas as pd
import joblib as jb
```

```python
model = jb.load('model.pkl')
x_scaler = jb.load('x_scaler.pkl')
```

```python
input=pd.DataFrame([[55,1,3,130,245,0,1,150,0,1.2,2,0,3]])
input_scld=x_scaler.transform(input)
prediction=model.predict(input_scld)
if prediction[0][0]>=0.5:
    print("Heart Diseases Detected")
else:
    print("No Heart Disease Detected")
```
